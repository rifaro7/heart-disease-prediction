Splitting the Features and Target

```python
X=heart_data.drop(columns='num',axis=1)
Y=heart_data['num']
print (X)
```

splitting the Data into Training data & Test data

Model Training


Logistic Regression

```python
model=LogisticRegression()
```

Converting categorical columns to numerical columns using one-hot encoding.

```python
X = pd.get_dummies(X, columns = ['sex', 'dataset', 'cp', 'restecg', 'exang', 'slope', 'thal'])
```

```python
print(X.head())
```

Splitting the encoded data into training and test sets.

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

Training the Logistic Regression model with the encoded training data.

```python
# Drop rows with NaN values from the training and test sets
X_train = X_train.dropna()
Y_train = Y_train[X_train.index] # Align Y_train with the dropped rows in X_train

X_test = X_test.dropna()
Y_test = Y_test[X_test.index] # Align Y_test with the dropped rows in X_test

model.fit(X_train, Y_train)
```

## Scalling the features



```python
from sklearn.preprocessing import StandardScaler

numerical_cols = X_train.select_dtypes(include=np.number).columns

scaler = StandardScaler()

X_train[numerical_cols] = scaler.fit_transform(X_train[numerical_cols])
X_test[numerical_cols] = scaler.transform(X_test[numerical_cols])
```

## Training the model

```python
model.fit(X_train, Y_train)
```

Model Evaluation


Accuracy Score


```python
#accuracy on training data
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on Training data :',training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
print('Accuracy on Test data :',test_data_accuracy)
```

Building a predictive system

```python
import pandas as pd

input_data = (55, 1, 3, 140, 250, 1, 2, 130, 1, 2.0, 1, 1, 3)

input_df = pd.DataFrame([input_data], columns=['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalch', 'exang', 'oldpeak', 'slope', 'ca', 'thal'])

categorical_cols = ['sex', 'cp', 'restecg', 'exang', 'slope', 'thal']

input_df_encoded = pd.get_dummies(input_df, columns=categorical_cols)

for col in X_train.columns:
    if col not in input_df_encoded.columns:
        input_df_encoded[col] = 0

input_df_encoded = input_df_encoded[X_train.columns]

numerical_cols_trained = X_train.select_dtypes(include=np.number).columns
input_df_encoded[numerical_cols_trained] = scaler.transform(input_df_encoded[numerical_cols_trained])

prediction = model.predict(input_df_encoded)
print(prediction)

if (prediction[0] == 0):
    print('The Person does not have a Heart Disease')
else:
    print('The Person has Heart Disease')
```
